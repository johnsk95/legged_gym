```python
import unittest
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch.autograd import Variable
import random
```

# Data

```python
if torch.cuda.is_available():
  device = torch.device('cuda')
else: 
  device = torch.device('cpu')
```

```python
# load data
d_np = pd.read_csv('data.csv').to_numpy()
# Normalize
# d_np = (d_np - d_np.min()) / (d_np.max() - d_np.min())
# d_np = d_np.to_numpy()

# idx = list(range(len(d_np)))
# shuffle data
# random.shuffle(idx)
```

```python
# df = pd.DataFrame(d_np)
# df.to_csv('train.csv', index=False)
```

```python
d_np = (d_np-d_np.min(axis=0))/((d_np.max(axis=0) - d_np.min(axis=0)))
X = d_np[:, :10]
y = d_np[:, 10:]

len = d_np.shape[0]
X_train = X[:int(len*0.9)]
X_test = X[int(len*0.9):]
y_train = y[:int(len*0.9)]
y_test = y[int(len*0.9):]

X_train = Variable(torch.Tensor(X_train)).to(device)
X_test = Variable(torch.Tensor(X_test)).to(device)
y_train = Variable(torch.Tensor(y_train)).to(device)
y_test = Variable(torch.Tensor(y_test)).to(device)
```

```python
print(d_np.shape)
print(X.shape)
print(y.shape)
print(X_train.size())
print(y_train.size())
print(X_test.size())
```

# Model

```python
class MLP(nn.Module):
  '''
    Multilayer Perceptron for regression.
  '''
  def __init__(self):
    super().__init__()
    self.layers = nn.Sequential(
      nn.Linear(10, 64),
      nn.ReLU(),
      nn.Linear(64, 64),
      nn.ReLU(),
      nn.Linear(64, 32),
      nn.ReLU(),
      nn.Linear(32, 3)
    )


  def forward(self, x):
    '''
      Forward pass
    '''
    return self.layers(x)
```

# Training

```python
num_epochs = 10
lr = 1e-4 #def 1e-4
window_size = 1

mlp = MLP().to(device)

loss_function = nn.L1Loss()

optimizer = torch.optim.Adam(mlp.parameters(), lr=lr)
```

```python
for epoch in range(num_epochs):
  # Set current loss value
  current_loss = 0.0
  
  # Iterate over the DataLoader for training data
  for i in range(window_size-1, X_train.size(0)):
    
    # Get and prepare inputs
    # inputs, targets = X_train[i], y_train[i]
    inputs, targets = X_train[i-window_size+1:i+1], y_train[i-window_size+1:i+1]
    inputs, targets = inputs.float(), targets.float()
    
    # Zero the gradients
    optimizer.zero_grad()
    
    # Perform forward pass
    outputs = mlp(inputs)
    
    # Compute loss
    loss = loss_function(outputs, targets)
    
    # Perform backward pass
    loss.backward()
    
    # Perform optimization
    optimizer.step()
    
    # # Print statistics
    # current_loss += loss.item()

  print(f'Epoch {epoch+1}, Loss {loss.item()}')

# Process is complete.
print('Training process has finished.')
```
